#  develop a Python program that is able to implement the gradient descent in order to achieve the linear regression (Single and Multivariables) of a set of datapoints.

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
from sklearn.metrics import r2_score 
```

#### Read RegData csv file into numpy array 

```python
data = np.genfromtxt('RegData - RegData.csv',delimiter=',')
data
```

```python
X=data[:,0]
Y=data[:,1]
```

#### Plot the original data (scatter plot of X,y)

```python
plt.scatter(X, Y)
plt.show()
```

## LR Full Implementation

```python
theta_0=0
theta_1=0
alpha=0.01
i=0
  
for i in range(0,2000):
    print("---------------iteration-----------",i)
    h=theta_0+theta_1*X
    print("h(x):", h)
    e=h-Y
    print("error",e) 
    print("\n")
    m = 2*len(X)
    total=(1/m)*sum(e**2)
    print("j",total)
    m2 =len(X)
    d_theta_0=(1/m2)*sum(e)
    d_theta_1=(1/m2)*sum((e)*X)
    G=np.array([[d_theta_0],
               [d_theta_1]])
    print("gradient vector:",G)
    print("\n")
    norm_g=np.linalg.norm(G)
    print("norm_gradient",norm_g)
    print("\n")
    if (norm_g<0.0454):
        break
    theta_0=theta_0 -  (alpha * d_theta_0)
    theta_1=theta_1 -  (alpha * d_theta_1)
    print(theta_0)
    print("\n")
    print(theta_1)
    print("\n")
    i+=1
print("--------------------training report--------------------")  
theta_0_opt=theta_0
theta_1_opt=theta_1
e=h-Y
print(e) 
print("\n")
m = 2*len(X)
total=(1/m)*sum(e**2)
print("\n cost",total)
print("\n y_pred:\n",h)
print("\n y_actual \n",Y) 

  
```

## Multivariate LR

```python
data2 = np.genfromtxt('/content/MultiVarLR - MultiVarLR (2).csv',delimiter=',')
```

```python
X2=data2[:,:-1]
```

```python
X2.shape
```

```python
X_0 = np.ones((25,1))
```

```python
x_new=np.hstack((X_0,X2))
y2=data2[:,3:4]
```

```python
theta = np.zeros((4,1))
alph =0.0001  
loss=[]
Theta=[]
H =[]
ch=0
i=0
G_norm = 1
while (i<5000):
    print(f"\n******* Iteration {i} *******")
    h_x = x_new @ theta
    H.append(h_x)
    print("h(x) = " ,h_x)

    err=(h_x - y2)
    m_2 =  2*len(x_new)
    j = 1/m_2 * err.T @ err
    loss.append(j[0][0])
    print("\nError Vector:\n" , err)
    print("\nj = " ,j)
    
    d_theta = (x_new.T @ err ) / len(x_new)
    print("Gradient Vector :\n",d_theta)
    G_norm = np.linalg.norm(d_theta)
    print("\nGradient Vector Norm :\n",G_norm)
    
    

    
    if G_norm <= 0.099 :
        break
    
    theta = theta -( alph *d_theta)
    Theta.append(theta)
    print("\ntheta_new : ",theta)
    

    i+=1
    
# ****** Training Report ********
print("\n******* Training Report *******")
print(f"\nGradient Descent converged after {i-1} iterations")
print("\ntheta_0_Opt : \n" ,theta)
print("\nError Vector: ",err)
print("\ncost : ",j)
print("\nh(x) = y_predict: \n" ,h_x)
print("\ny_actual:\n" , y2)
print()
```
