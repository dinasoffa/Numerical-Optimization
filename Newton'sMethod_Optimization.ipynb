```python
Df_1 = lambda X: np.array([2*X[0],2*X[1]])
DDf_1 = lambda X: np.array([[2,0],
                         [0,2]])
```

```python
f_2 = lambda x: x**2 + x*y + y**2
Df_2 = lambda X: np.array([2*X[0]+X[1],2*X[1]+X[0]])
DDf_2 = lambda X: np.array([[2,1],
                        [1,2]])
```

```python
def Gradient_Multi(x0, max_it,alpha ,epsilon, df ):
    x=x0
     
    for i in range(max_it):
        print(f"\n******* Iteration {i} *******")
        x = x -  (df(x) *alpha)
        print("x =  ",x)
        grad = np.abs(df(x))
        print("grad=",grad)
        if grad[0] < epsilon and grad[1] < epsilon:
            break
        
     
    
    return x , grad ,i
```

```python
x,grad,i=Gradient_Multi((10,10),200,0.1,0.001,Df_1)
```

```python
x,grad,i=Gradient_Multi((10,10),200,0.5,0.001,Df_1)
```

```python
def newton_Multi(x0, max_it ,epsilon, df,DDf_1,alpha ):
    x=x0
     
    for i in range(max_it):
        print(f"\n******* Iteration {i} *******")
        x = x -  alpha*(np.linalg.inv(DDf_1(x)))*df(x)
        grad = np.abs(df(x))
        if grad[0,0] < epsilon and grad[1,1] < epsilon:
            break
        
    x_new=np.array([x[0,0],x[1,1]])
    gradient=np.array([grad[1,1],grad[1,1]])

    print(x_new)
    print(gradient)
    
```

```python
newton_Multi((10,10),200,0.001,Df_1,DDf_1,1)
```
